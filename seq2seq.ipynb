```python
import random
import numpy as np
from keras import layers
from keras.models import Sequential
from keras.layers.core import Dense, Reshape
from keras.layers.wrappers import TimeDistributed
```

```python
# パラメータ設定

# データ次元
DIM = 30

# 入力シーケンスの最大長
INPUT_MAX_LEN = 10

# 出力シーケンスの最大長
OUTPUT_MAX_LEN = 10

# 学習データ数
TRAINING_SIZE = 200
```

```python
# データ生成
x_data = []
y_data = []

for i in range(TRAINING_SIZE + 10):
    # 入力の累積を出力
    s = 0
    xx = []
    yy = []
    for j in range(INPUT_MAX_LEN):
        rnd = random.randint( 0, 3 )
        x = np.zeros( DIM, dtype=np.bool )
        x[rnd] = True   # one-hot voctorにする
        s += rnd
        
        y = np.zeros( DIM, dtype=np.bool )
        y[s] = True   # one-hot voctorにする

        xx.append(x)
        yy.append(y)
        
    x_data.append(xx)
    y_data.append(yy)
        
x_data = np.array(x_data)
y_data = np.array(y_data)


# 学習用と評価用に分割
x_train = x_data[:TRAINING_SIZE]
y_train = y_data[:TRAINING_SIZE]

x_eval = x_data[TRAINING_SIZE:]
y_eval = y_data[TRAINING_SIZE:]
```

```python
# モデル構築
HIDDEN_SIZE = 256
BATCH_SIZE = 128
LAYERS = 1

model = Sequential()
model.add(layers.LSTM(HIDDEN_SIZE, input_shape=(INPUT_MAX_LEN, DIM)))
model.add(Dense(OUTPUT_MAX_LEN * DIM))
model.add(Reshape((OUTPUT_MAX_LEN, DIM)))
model.add(TimeDistributed(Dense(DIM, activation='softmax')))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# 学習
for iteration in range(1, 200):
    print()
    print('-----------')
    print('Iteration', iteration)
    model.fit(x_train, y_train, batch_size=BATCH_SIZE, epochs=1, validation_data=(x_eval, y_eval))
```

```python
# テスト
for i in range(10):
    preds = model.predict_classes( np.array([x_eval[i]]), verbose=0)
    print( "input: ",  [ np.argmax(x) for x in x_eval[i] ] )
    print( "output: ", preds )
    print("--------")
```
